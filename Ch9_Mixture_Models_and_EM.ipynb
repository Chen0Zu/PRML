```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
```

```python
x1 = np.random.multivariate_normal([1,1], .4*np.eye(2), 100)
x2 = np.random.multivariate_normal([-1,-1], .4*np.eye(2), 100)
x_train = np.r_[x1, x2]
y_train = np.r_[np.zeros([100,1]), np.ones([100,1])]

plt.figure()
plt.scatter(x_train[:,0], x_train[:,1], c=y_train.flatten())
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.gca().set_aspect("equal", adjustable='box')

max_iter = 6
k = 2
center = np.array([[-2.0, 2.0], [2.0, -2.0]])
plt.figure(figsize=[10,8])
J = np.zeros([2*max_iter, 1])
for i in range(max_iter):
#     E-Step
    pair_dist = cdist(x_train, center)**2
    cluster = np.argmin(pair_dist, axis=1).reshape(-1,1)
    J[i] = np.sum(pair_dist[cluster.flatten() == 0, 0]) + np.sum(pair_dist[cluster.flatten() == 1, 1])
#     M-Step
    for j in range(k):
        center[j,:] = np.mean(x_train[(cluster == j).flatten(), :], axis=0, keepdims=True)
    J[i+1] = np.sum(pair_dist[cluster.flatten() == 0, 0]) + np.sum(pair_dist[cluster.flatten() == 1, 1])
    
    plt.subplot(2, 3, i+1)
    plt.scatter(x_train[:, 0], x_train[:, 1], c=cluster.flatten())
    plt.scatter(center[:, 0], center[:, 1], s = 200, marker='x', linewidths=2, c = ['red', 'blue'], edgecolors='white')
    plt.xlim(-3,3)
    plt.ylim(-3,3)
    plt.gca().set_aspect("equal", adjustable='box')

plt.figure()
plt.plot(np.arange(max_iter*2), J, marker = 'o')
```
