```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
```

```python
x1 = np.random.multivariate_normal([1,1], .4*np.eye(2), 100)
x2 = np.random.multivariate_normal([-1,-1], .4*np.eye(2), 100)
x_train = np.r_[x1, x2]
y_train = np.r_[np.zeros([100,1]), np.ones([100,1])]

plt.figure()
plt.scatter(x_train[:,0], x_train[:,1], c=y_train.flatten())
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.gca().set_aspect("equal", adjustable='box')

max_iter = 6
k = 2
center = np.array([[-2.0, 2.0], [2.0, -2.0]])
plt.figure(figsize=[10,8])
J = np.zeros([2*max_iter, 1])
for i in range(max_iter):
#     E-Step
    pair_dist = cdist(x_train, center)**2
    cluster = np.argmin(pair_dist, axis=1).reshape(-1,1)
    J[i] = np.sum(pair_dist[cluster.flatten() == 0, 0]) + np.sum(pair_dist[cluster.flatten() == 1, 1])
#     M-Step
    for j in range(k):
        center[j,:] = np.mean(x_train[(cluster == j).flatten(), :], axis=0, keepdims=True)
    J[i+1] = np.sum(pair_dist[cluster.flatten() == 0, 0]) + np.sum(pair_dist[cluster.flatten() == 1, 1])
    
    plt.subplot(2, 3, i+1)
    plt.scatter(x_train[:, 0], x_train[:, 1], c=cluster.flatten())
    plt.scatter(center[:, 0], center[:, 1], s = 200, marker='x', linewidths=2, c = ['red', 'blue'], edgecolors='white')
    plt.xlim(-3,3)
    plt.ylim(-3,3)
    plt.gca().set_aspect("equal", adjustable='box')

plt.figure()
plt.plot(np.arange(max_iter*2), J, marker = 'o')
```

# 9.2.2 EM for Gaussian mixtures

```python
def gaussian_pdf(x, mu, sigma):
    p = 1/(2*np.pi*np.linalg.det(sigma)**.5)*np.exp(-0.5*(x-mu)@np.linalg.inv(sigma)@(x-mu).T)
    return np.diag(p).reshape(-1,1)

np.random.seed(1)
x1 = np.random.multivariate_normal([1,1], .2*np.eye(2), 100)
x2 = np.random.multivariate_normal([-1,-1], .2*np.eye(2), 100)
x_train = np.r_[x1,x2]
y_train = np.r_[np.zeros([100,1]), np.ones([100,1])]

#x0, x1 = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
#x = np.array([x0, x1]).reshape(2, -1).T
#
#plt.contour(x0,x1,gaussian_pdf(x,[-1,1], np.eye(2)).reshape(100,100))
#plt.xlim(-3,3)
#plt.ylim(-3,3)



plt.figure()
plt.scatter(x_train[:,0], x_train[:,1], marker='o')
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.gca().set_aspect('equal', adjustable='box')

means = np.array([[-2.0,2.0],[2.0,-2.0]])
covs = np.zeros([2,2,2])
for i in range(2):
    covs[i] = np.eye(2)
k = 2
N = 200
Nk = np.zeros([k,1])
pi = np.array([0.6,0.4])
gamma = np.zeros([N,k])
z = np.zeros([N,k])
max_iter = 100
#plt.figure()
for i in range(max_iter):
#     E-step
    for j in range(k):
        z[:,j] = gaussian_pdf(x_train, means[j], covs[j]).flatten()
        gamma[:,j] = pi[j]*gaussian_pdf(x_train, means[j], covs[j]).flatten()
    gamma /= gamma.sum(axis=1,keepdims=True)
#     M-step
    Nk = gamma.sum(axis=0)
    for j in range(k):
        means[j] = 1/Nk[j]*np.sum(gamma[:,j].reshape(-1,1)*x_train,axis=0)
        covs[j] = 1/Nk[j]* (x_train - means[j]).T @ np.diag(gamma[:,j]) @ (x_train - means[j])
        pi[j] = Nk[j]/N
    if(i % 10 == 0):
#        plt.subplot(max_iter,1,i+1)
        plt.figure()
        plt.scatter(x_train[:,0], x_train[:,1], marker='o', c = gamma[:,0])
        plt.scatter(means[:,0], means[:,1], marker='X', s=200, c=['red', 'green'], edgecolors='white', lw=2)


plt.figure()
plt.scatter(x_train[:,0], x_train[:,1], marker='o')
plt.scatter(means[:,0], means[:,1], marker='X', s=200, c=['red', 'green'], edgecolors='white', lw=2)
```
