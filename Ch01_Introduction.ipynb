# 1.1 Polynomial Curve Fitting

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100).reshape(-1,1)
t = np.sin(2*np.pi*x)
plt.figure()
plt.plot(x,t, c = 'g')
plt.xlabel('x')
plt.ylabel('t')

N = 10
np.random.seed(3)
x_observed = np.random.uniform(0,1,N).reshape(-1,1)
y_observed = np.sin(2*np.pi*x_observed) + np.random.normal(loc=0, scale = 0.1, size = N).reshape(-1,1)
plt.scatter(x_observed,y_observed, marker = 'o', c='', edgecolors='b', s = 100)
```

```python
def polynomial(x, M):
    n = x.shape[0]
    x_poly = np.zeros([n, M+1])
    for i in range(M+1):
        x_poly[:,i] = x.reshape(-1)**i
    return x_poly

z = np.array([1,2,3]).reshape(-1,1)
polynomial(z,3)
```

```python
x_train = x_observed
y_train = y_observed

X = polynomial(x_train,0)
w0 = np.linalg.pinv(X.T@X)@X.T@y_train
y0 = polynomial(x,0)@w0

X = polynomial(x_train,1)
w1 = np.linalg.pinv(X.T@X)@X.T@y_train
y1 = polynomial(x,1)@w1

X = polynomial(x_train,3)
w3 = np.linalg.pinv(X.T@X)@X.T@y_train
y3 = polynomial(x,3)@w3

X = polynomial(x_train,9)
w9 = np.linalg.pinv(X.T@X)@X.T@y_train
y9 = polynomial(x,9)@w9

plt.figure()
plt.subplot(2,2,1)
plt.plot(x,t,c='g')
plt.scatter(x_train,y_train,marker='o',c='',edgecolors='b',s=70)
plt.plot(x,y0,c='r')
plt.title('M=0')

plt.subplot(2,2,2)
plt.plot(x,t,c='g')
plt.scatter(x_train,y_train,marker='o',c='',edgecolors='b',s=70)
plt.plot(x,y1,c='r')
plt.title('M=1')

plt.subplot(2,2,3)
plt.plot(x,t,c='g')
plt.scatter(x_train,y_train,marker='o',c='',edgecolors='b',s=70)
plt.plot(x,y3,c='r')
plt.title('M=3')

plt.subplot(2,2,4)
plt.plot(x,t,c='g')
plt.scatter(x_train,y_train,marker='o',c='',edgecolors='b',s=70)
plt.plot(x,y9,c='r')
plt.title('M=9')
plt.ylim([-1.2,1.2])

plt.subplots_adjust(left=None,bottom=None, right=None, top = None, wspace=None, hspace=0.4)
```

```python
x_test = np.random.uniform(0,1,100).reshape(-1,1)
y_test = np.sign(2*np.pi*x_test) + np.random.normal(0,0.1,100).reshape(-1,1)
M = 10
mse_training = np.zeros([M,1])
mse_test = np.zeros([M,1])
for i in range(M):
    X = polynomial(x_train,i)
    w = np.linalg.pinv(X.T@X)@X.T@y_train
    
    mse_training[i] = np.linalg.norm(X@w-y_train)/np.sqrt(X.shape[0])
    mse_test[i] = np.linalg.norm(polynomial(x_test,i)@w - y_test) / np.sqrt(x_test.shape[0])

plt.figure()
plt.plot(range(M), mse_training, marker='o')
plt.plot(range(M), mse_test, marker='o')
plt.ylim([-0.1,2])
plt.legend(['Training', 'Test'])
plt.xlabel('M')
plt.ylabel('E_rms')
```

# 1.2 Probability Theory

```python
mu = 0
sigma = 2
x = np.linspace(-10, 10, 100).reshape(-1,1)
y = 1/(2*np.pi*sigma**2)**0.5*np.exp(-0.5*(x-mu)**2/sigma**2)
plt.figure()
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('$p(x)')
plt.title('Gaussian Distribution ($\mu=0,\sigma=2$)')
```

```python
M = 9
alpha = 5e-3
beta = 10
np.random.seed(3)
x_observed = np.random.uniform(0,1,N).reshape(-1,1)
y_observed = np.sin(2*np.pi*x_observed) + np.random.normal(loc=0, scale = 0.1, size = N).reshape(-1,1)
x = np.linspace(0, 1, 100).reshape(-1,1)
t = np.sin(2*np.pi*x)
x_train = x_observed
y_train = y_observed
x_test = x
y_test = t
X_train = polynomial(x_train, M)
X_test = polynomial(x_test, M)

n_train = x_train.shape[0]
n_test = x_test.shape[0]
y_pred = np.zeros([n_test, 1])
std_pred = np.zeros([n_test, 1])

inv_S = alpha*np.eye(M+1) + beta*X_train.T@X_train
for i in range(n_test):
    x = X_test[i, :].reshape(-1,1)
    y_pred[i] = beta * x.T@np.linalg.inv(inv_S) @ X_train.T @ y_train
    std_pred[i] = 1/beta + x.T@np.linalg.inv(inv_S)@x
    
plt.figure()
plt.plot(x_test, y_test, c = 'g')
plt.scatter(x_train, y_train, c='', edgecolors='b',s=70, marker='o')
plt.plot(x_test, y_pred, c='r')
plt.fill_between(x_test.flatten(), y_pred.flatten()-np.sqrt(std_pred.flatten()), y_pred.flatten()+np.sqrt(std_pred.flatten()), alpha = 0.5)
plt.annotate('M=9', xy=(0.5,1))
plt.ylim([-1.2, 1.2])
plt.xlabel('x')
plt.ylabel('t')
plt.legend(['sin$(2\pi x)$', 'training data', 'prediction', 'std'])
```

# 1.4 The Curse of Dimensionality

```python
def volume_of_ball(n, r):
    volume = 0
    if n % 2 == 0:
        volume = np.pi**(n/2)*r**n/np.math.factorial(int(n/2))
    else:
        volume = 2**n*np.math.factorial(int((n-1)/2))/np.math.factorial(n)*np.pi**(int((n-1)/2))*r**n
    return volume

def volume_of_cube(n,r):
    volume = (2*r)**n
    return volume

n = 20
d = np.linspace(1, n, n)
ratio = np.zeros([n,1])
for i in range(n):
    ratio[i] = volume_of_ball(d[i], r=1)/volume_of_cube(d[i], r=1)
    
plt.figure()
plt.plot(d, ratio, marker='o',ls='-',markersize=10,markeredgecolor='b',markerfacecolor='w')
plt.xlabel('Dimension')
plt.ylabel('Volume fraction')
plt.title('Fraction of the volume of a sphere lying in a cube')
```

# 1.6 Information Theory

```python
def entropy(x):
    c = np.unique(x)
    c_n = c.shape[0]
    p = np.zeros([c_n, 1])
    n = x.shape[0]
    for i in range(c_n):
        p[i] = sum(x == c[i]) / n
    H = -p.T@np.log2(p)
    return H, p

def relative_entropy(x, y):
    dump, px = entropy(x)
    dump, py = entropy(y)
    kl = -px.T@np.log2(py) + px.T@np.log(px)
    return kl

def cross_entropy(x, y):
    dump, px = entropy(x)
    dump, py = entropy(y)
    crossH = -px.T@np.log2(py)
    return crossH
```
